```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


df = pd.read_csv('C:/Users/PC/Desktop/M.python/ecommerce_estatistica.csv')
print(df.head().to_string())


# GRÁFICO HISTOGRAMA
plt.figure(figsize = (10, 6))
plt.hist(df['Preço'], bins=40, color='skyblue', alpha=0.8)
plt.title('1. Histograma: Distribuição de Preços dos Produtos')
plt.xlabel('Preço (R$)')
plt.ylabel('Frequência')
plt.grid(True)
plt.axvline(df['Preço'].median(), color='skyblue', label=f'Mediana:{df['Preço'].median():.2f}')
plt.show()
```

```python
#  GRÁFICO DE DISPERSÃO

sns.scatterplot(x='Preço', y='Nota', data=df, alpha=0.6, color='blue')
plt.title('2. Dispersão: Preço vs. Nota de Avaliação', fontsize=18)
plt.xlabel('Preço (R$)')
plt.ylabel('Nota Média (1-5)')
```

```python
# GRÁFICO MAPA DE CALOR

corr = df[['Nota', 'N_Avaliações']].corr()
plt.subplot(2, 2, 3)
sns.heatmap(corr, annot=True, cmap='coolwarm', linecolor='black')
plt.title('3. Mapa de Calor da Correlação de Preço & Nota de Avaliação', fontsize=18)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de Barra

plt.figure(figsize=(10, 6))
sns.barplot(x='Temporada', y='Nota', data=df, color='skyblue')
plt.title('4. Gráfico de Barras: Nota Média por Temporada', fontsize=16)
plt.xlabel('Temporada', fontsize=12)
plt.ylabel('Nota Média de Avaliação', fontsize=12)
plt.ylim(4.0, 5.0)
plt.xticks(rotation=0)
plt.show()
```
