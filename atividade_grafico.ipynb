```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


df = pd.read_csv('C:/Users/PC/Desktop/M.python/ecommerce_estatistica.csv')
print(df.head().to_string())


# GR√ÅFICO HISTOGRAMA
plt.figure(figsize = (10, 6))
plt.hist(df['Pre√ßo'], bins=40, color='skyblue', alpha=0.8)
plt.title('1. Histograma: Distribui√ß√£o de Pre√ßos dos Produtos')
plt.xlabel('Pre√ßo (R$)')
plt.ylabel('Frequ√™ncia')
plt.grid(True)
plt.axvline(df['Pre√ßo'].median(), color='skyblue', label=f'Mediana:{df['Pre√ßo'].median():.2f}')
plt.show()
```

```python
#  GR√ÅFICO DE DISPERS√ÉO

sns.scatterplot(x='Pre√ßo', y='Nota', data=df, alpha=0.6, color='blue')
plt.title('2. Dispers√£o: Pre√ßo vs. Nota de Avalia√ß√£o', fontsize=18)
plt.xlabel('Pre√ßo (R$)')
plt.ylabel('Nota M√©dia (1-5)')
```

```python
# GR√ÅFICO MAPA DE CALOR

corr = df[['Nota', 'N_Avalia√ß√µes']].corr()
plt.subplot(2, 2, 3)
sns.heatmap(corr, annot=True, cmap='coolwarm', linecolor='black')
plt.title('3. Mapa de Calor da Correla√ß√£o de Pre√ßo & Nota de Avalia√ß√£o', fontsize=18)
plt.tight_layout()
plt.show()
```
